```python
from pathlib import Path
import pandas as pd

LOSS_VALUES_FRILENAME = "epoch_{}.csv"
root = Path("/home/azureuser/localfiles/cpath/outputs/2022-09-27T143238Z_TP53TilesTcgaPrad/loss_values")
n_epochs = 10
epochs = list(range(n_epochs))
limit = 5
slides_with_highest_loss_values = []
highest_loss_values = []
for epoch in epochs:
    loss_values = pd.read_csv(root / LOSS_VALUES_FRILENAME.format(epoch))
    slides_with_highest_loss_values.append(loss_values["slide_id"][:limit])
    highest_loss_values.append(loss_values["loss"][:limit])
```

```python
import matplotlib.pyplot as plt
limit = 5
plt.figure(figsize=(10, 10))
for i in range(limit-1, -1, -1):
    epochs
    slides = [slides_with_highest_loss_values[j][i] for j in epochs]
    loss = [highest_loss_values[j][i] for j in epochs]
    plt.scatter(epochs, slides)
    coordinates = [(l, _x, _slide) for l, _x, _slide in zip(loss, epochs, slides)]
    for x in coordinates: plt.annotate(f"{x[0]:.3f}", (x[1], x[2]))
    plt.xlabel("Epoch")
    plt.ylabel("Slide ID")
plt.title("Slides with highest loss values per epoch")
plt.show()
```

```python
epoch = 0
loss_cache = pd.read_csv(root / LOSS_VALUES_FRILENAME.format(epoch))
slides = loss_cache["slide_id"]
```

```python
slides
```

```python

loss_val = {slide_id: [] for slide_id in slides}
for epoch in epochs:
    loss_cache = pd.read_csv(root / LOSS_VALUES_FRILENAME.format(epoch))
    loss_cache.set_index("slide_id", inplace=True)
    loss_val = {slide_id: loss_val[slide_id] + [loss_cache.loc[slide_id, "loss"]] for slide_id in slides}
```

```python
import numpy as np
loss_values = []
for slide_id in slides:
    loss_values.append(loss_val[slide_id])
loss_values = np.array(loss_values)
loss_values.shape
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt

plt.figure(figsize=(15, 15))
ax = sns.heatmap(loss_values, linewidth=0.5, annot=True, yticklabels=slides)
plt.xlabel("Epoch")
plt.ylabel("Slide ID")
plt.show()
```
